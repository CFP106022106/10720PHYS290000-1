```python
import math
import numpy as np
import matplotlib.pyplot as plt

pigx=np.random.random(1)*100
pigy=np.random.random(1)*100
plt.plot(pigx,pigy,'r.')
plt.axis([0, 100, 0, 100])
```

```python
px=np.rint(pigx)
py=np.rint(pigy)
theta=30
vo=40
h=theta*math.pi/180
g=9.8
t = np.arange(0,10,0.01)

plt.plot(pigx,pigy,'r.')

x,y=vo*t*math.cos(h),vo*t*math.sin(h)-0.5*g*t**2

for t in (0,10,0.01):
    plt.plot(x,y)
    plt.axis([0, 100, 0, 100])
    ix,iy=np.rint(x),np.rint(y)
   
    if  any(ix==px) and any(iy==py):
        print('耶')
      


```
